## **Preparation Data**

```python
# movies drop tag null
movies = movies.dropna(subset=['tag'])
```

```python
# drop timestamp ratings
ratings = ratings.drop('timestamp', axis=1)
```

```python
# join dataset using left join
df = pd.merge(movies[['movieId', 'title', 'genres', 'tag', 'year']], ratings[['movieId','rating']], on='movieId')
```

```python
df
```

```python
# Because it was felt that there was too much data, the average rating was made into one line
# Change ratings based on average
df = df.groupby(['movieId', 'title', 'genres','tag', 'year'], as_index=False)['rating'].mean()
```

```python
df
```

```python
# modify 'rating' column to display only two decimal
df['rating'] = df['rating'].apply(lambda x: f"{x:.2f}")
```

```python
df
```

```python
df_memory = df.memory_usage()
df_memory
```

```python
# for saving memory, convert movieId column from int64 to int32
df.movieId = df.movieId.astype('int32')
```

```python
# for saving memory, convert year column from int64 to int32
df.year = df.year.astype('int32')
```

```python
# Drop genres values (no genres listed) in genres column
df = df[~df['genres'].str.contains('(no genres listed)', na=False)]
```

```python
df
```

```python
#Filter data by year
data_filter = df[(df['year'] >= 1995) & (df['year'] <= 2019)]
```

```python
# Delete data that doesn't pass the filter
df.drop(df[~df['year'].isin(data_filter['year'])].index, inplace=True)
```

```python
df
```

```python
import re
#Cleaning tag column
def clean_tag(tag):
  # Hapus simbol - ! dan simbol lain
  tag = re.sub(r"[^\w, ]", "", tag)
  
  # Kembalikan tag yang sudah dibersihkan
  return tag

df['tag'] = df['tag'].apply(clean_tag)
```

```python
# convert genres to lowercase 
df['genres'] = df['genres'].str.lower()
```

```python
# convert tag to lowercase
df['tag'] = df['tag'].str.lower()
```
